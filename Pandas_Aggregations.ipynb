# Group By
# CrossTab
# Pivot Table

```python
import pandas as pd
import numpy as np
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Sample - Superstore.xls")
df.head()
```

```python
df.columns
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
df.groupby('Category')[['Sales','Profit']].sum()
```

```python
df.groupby(['Category','Region'])['Sales'].sum()
```

```python
df.groupby(['Category','Region'])['Sales'].describe().T
```

```python
df.groupby(['Category','Region'])['Sales'].sum().sort_values(ascending=False)
```

```python
pd.crosstab(df['Region'],df['Category'])
```

```python
pd.pivot_table(df,values='Sales',columns='Category',index='Region',aggfunc='sum')
```

```python
pd.pivot_table(df,values=['Sales','Quantity'],index='Category',columns='Region',aggfunc='sum')
```
